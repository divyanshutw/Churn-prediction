```python
x=df.drop(["churn"],axis=1)
y=df['churn']
x.shape,y.shape
```

```python
x=pd.get_dummies(x)
x
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=10,stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_x=scaler.fit_transform(x)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn import metrics
clf=knn()
```

```python
train_x.isnull().sum()
```

```python
train_x.shape,train_y.shape
```

```python
test_error=[]
for k in range(40,60,2):
    clf=knn(k)
    #clf.fit(x,y)
    clf.fit(train_x,train_y)
    temp=clf.predict(test_x)
    er=metrics.f1_score(temp,test_y)
    er=1-er
    test_error.append(er)
test_error
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(train_x,train_y)
temp=lr.predict(test_x)
```

```python
test1=metrics.f1_score(temp,test_y)
test2=metrics.mean_absolute_error(test_y,temp)
test1,test2
```
